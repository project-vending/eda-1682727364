````
# Importing necessary libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```


````

````
#
````

````
# Loading the dataset into pandas dataframe

``` python
````

````
# loading dataset into dataframe
data = pd.read_csv('https://raw.githubusercontent.com/midhunmohank/INFO6105/main/advertising.csv')
```

````

````
#
````

````
# Checking for null values

``` python
````

````
# checking for nulls
data.isnull().sum()
```

````

````
#
````

````
# Describing the dataset

``` python
````

````
# describing the dataset
data.describe()
```

````

````
#
````

````
# Plotting the distribution of different variables

``` python
````

````
# plotting distribution of different variables
plt.figure(figsize=(10,6))
plt.title("Distribution Plot")
sns.histplot(data=data, x="TV", kde=True)
sns.histplot(data=data, x="Radio", kde=True)
sns.histplot(data=data, x="Newspaper", kde=True)
plt.show()
```

````

````
#
````

````
# Plotting the boxplot of different variables

``` python
````

````
# plotting boxplot of different variables
plt.figure(figsize=(10,6))
sns.boxplot(data=data)
plt.show()
```

````

````
#
````

````
# Plotting the correlation heatmap

``` python
````

````
# plotting correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.show()
```
````
